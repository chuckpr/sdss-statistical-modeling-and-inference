## Jupyter Lab Interface

Jupyter provides a rich coding environment for data scientists. The interface enables you to write code directly in the browser and execute it. The feedback is available in the output below your code. Output can be simple text, tables, and even data vizualizatoins.

Here were load in some an R packages...

```r
library('palmerpenguins')
library('tidyverse')
```

We can look at a table of data...

```r
penguins |> head()
```

...and visualize the data!

```r
options(repr.plot.width = 6, repr.plot.height = 4, repr.plot.res = 300)
theme_set(theme_gray(base_size = 16))

p = ggplot(penguins, aes(x = bill_depth_mm, y = flipper_length_mm, color = species)) + 
    geom_point(na.rm = T) +
    scale_color_brewer(palette = 'Set2')

p
```
